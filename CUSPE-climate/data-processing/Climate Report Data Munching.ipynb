```python
import numpy as np
import pandas as pd
```

```python
data = []

with open('./data/csvs/carbon-footprint-data.txt','r') as file:
    for line in file:
        splitline = []
        for point in line.split('\t'):
            splitline.append(point)
        data.append(splitline)
```

```python
data
```

```python
data_array = np.zeros((len(data),len(data[0])))
data_array = data_array.astype(dtype=object)
print(data_array.shape)
```

```python
for i, line in enumerate(data):
    for j, point in enumerate(line):
        data_array[i,j] = line[j]
```

```python
data_array
```

## Some data fixing

```python
#data_array[0,:]
for i, point in  enumerate(data_array[:,-1]):
    try: 
        data_array[i,-1] = point[:-1]
    except TypeError:
        continue
```

```python
for i, point in  enumerate(data_array[1:,7]):
    if point == '' or point == 'N/A':
        data_array[i+1,7] = 'N/A'
        
    else:
        try:
            data_array[i+1,7] = float(point)
        except ValueError:
            data_array[i+1,7] = 'N/A'
```

```python
data_array[1:,5]
```

```python
float(data_array[1,5].replace('"','').replace(',',''))
```

```python
for i, point in  enumerate(data_array[1:,5]):
    if point == '' or point == 'N/A' or point == ' -   ':
        data_array[i+1,5] = 'N/A'
    else:
        data_array[i+1,5] = float(str(point).replace('"','').replace(',',''))
```

```python
for i, point in  enumerate(data_array[1:,8]):
    if point == '' or point == 'N/A' or point == ' -   ':
        data_array[i+1,8] = 'N/A'
        
    else:
        data_array[i+1,8] = float(str(point).replace('"','').replace(',',''))
```
