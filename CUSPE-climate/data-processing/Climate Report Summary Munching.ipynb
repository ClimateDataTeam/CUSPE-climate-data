```python
import numpy as np
import pandas as pd
```

```python
#this is likely easiest done manually
```

```python
data = []

with open('./data/csvs/gross summary.txt','r') as file:
    for line in file:
        splitline = []
        for point in line.split('\t'):
            splitline.append(point)
        data.append(splitline)
```

```python
data_array = np.zeros((len(data),len(data[0])))
data_array = data_array.astype(dtype=object)
print(data_array.shape)
```

```python
for i, line in enumerate(data):
    for j, point in enumerate(line):
        data_array[i,j] = line[j]
```

```python
data_array
```

```python
#data_array[0,:]
for i, point in  enumerate(data_array[:,-1]):
    try: 
        data_array[i,-1] = point[:-1]
    except TypeError:
        continue
```

```python
df = pd.DataFrame(data_array[1:,:].T, index=data_array[0])
df
```

```python
data_array[1:,4]
```

```python
for column in range(1,6):
    for i, point in  enumerate(data_array[1:,column]):
        if point == '' or point == 'N/A' or point == ' -   ':
            data_array[i+1,column] = 0
        else:
            data_array[i+1,column] = float(str(point).replace('"','').replace(',',''))
```

```python
df = pd.DataFrame(data_array[1:,:].T, index=data_array[0])
df
```

```python
df.to_pickle('./data/gross-summary.pkl')
```
